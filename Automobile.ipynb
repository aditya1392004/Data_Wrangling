```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing data
data= ("auto.csv")
df= pd.read_csv(data)
df.head()
```

```python
df.columns
```

```python
headings= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm", "city-mpg","highway-mpg","price"]
df.columns= headings
```

```python
df.head()
```

# 

# Replacing null values

```python
df= df.replace('?', np.nan)
```

```python
missing= df.isnull()
missing.head()
```

```python
df['normalized-losses'].astype('float64').mean()
```

```python
df['normalized-losses'].replace(np.nan, 122, inplace= True)
```

```python
df.head()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df['bore'].astype('float64').mean()
```

```python
df['bore'].replace(np.nan, 3.33, inplace= True)
```

```python
df['stroke'].astype('float64').mean()
```

```python
df['stroke'].replace(np.nan, 3.26, inplace=True)
```

```python
df['horsepower'].astype('float64').mean()
```

```python
df['horsepower'].replace(np.nan,104.22,inplace=True)
```

```python
df['peak-rpm'].astype('float64').mean()
```

```python
df['peak-rpm'].replace(np.nan,5125.99,inplace=True)
```

#

# Dropping rows in which prices are not mentioned

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)    #Resetting index
```

# 

# Changing datatypes

```python
df.dtypes
```

```python
df[['normalized-losses','horsepower','peak-rpm','price']] = df[['normalized-losses','horsepower','peak-rpm','price']].astype('int64')
df[['bore','stroke']] = df[['bore','stroke']].astype('float64')
```

```python
df[['length', 'width', 'height']]
```

#

# Standardization

```python
df['city-L/100km'] = 235/df["city-mpg"]
df['highway-L/100km'] = 235/df['highway-mpg']
df.head()
```

#

# Normalization

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df.head()
```

```python
df[['length', 'width', 'height']]
```

# 

# Final analysis

```python
plt.hist(df['horsepower'])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins

group_names = ['Low', 'Medium', 'High']

df['horsepower-binned'] = pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-binned']].head(10)
```

```python
df['horsepower-binned'].value_counts()
```

```python
plt.bar(group_names,df['horsepower-binned'].value_counts())
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
df = pd.concat([df,dummy_variable_1],axis=1)
df.drop('fuel-type',axis=1,inplace=True)
df.head()
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.head()
df = pd.concat([df,dummy_variable_2],axis=1)
df.drop('aspiration',axis=1,inplace=True)
df.head()
```

```python
sb.regplot(x='price', y='engine-size',data=df)
plt.show()
```

```python
sb.regplot(x='price',y='highway-mpg',data=df)
plt.show()
```

```python
sb.boxplot(x='body-style',y='price',data=df)
plt.show()
```

```python
sb.boxplot(x='engine-location',y='price',data=df)
plt.show()
```

```python
grouped = df.groupby(['drive-wheels','body-style'])['price'].mean()
grouped = grouped.reset_index()
grouped
```

```python
grouped_pivot = grouped.pivot('drive-wheels','body-style','price')
grouped_pivot
```

```python
plt.pcolor(grouped_pivot,cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot,cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.astype(str).tolist()
col_labels = grouped_pivot.index

ax.set_xticks(np.arange(grouped_pivot.shape[1])+.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0])+.5, minor=False)

ax.set_xticklabels(row_labels, minor = False)
ax.set_yticklabels(col_labels, minor = False)

plt.xticks(rotation = 90)

fig.colorbar(im)

plt.show()
```
